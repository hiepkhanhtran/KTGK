# Kiểm tra giữa kỳ nhập môn phân tích dữ liệu và học sâu

```python
#1.	Tạo Dataframe dữ liệu đã cho
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Tạo DataFrame như dữ liệu đã cho trên.
data = {
    'Name': ['Mr. Harris', 'Mrs. Bradley', 'Miss. Laina', 'Mrs. Heath', 'Mr. William', 'Mr. James', 
             'Mr. Timothy', 'Mrs. Oscar', 'Mrs. Nicholas', 'Miss. Marguerite', 'Mr. William', 'Miss. Hulda', 
             'Mr. Lawrence', 'Mr. Wilam'],
    'Height': [1.80, 1.61, 1.78, 1.74, 1.55, 1.90, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    'Weight': [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    'AGE': [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)
```

```python
#2.Hiển thị thông tin Dataframe vừa tạo.
print(df)
```

```python
#3.Tính giá trị trung bình của cột <AGE>.
mean_age = df['AGE'].mean()
print(f"Trung bình AGE: {mean_age}")
```

```python
#4.Nhóm dữ liệu cột <Height> và tính trung bình cột <Weight>.
grouped_data = df.groupby('Height')['Weight'].mean().reset_index()
print(grouped_data)
```

```python
#5. Sắp xếp DataFrame theo cột <Weight> tăng dần.
sorted_df = df.sort_values('Weight')
print(sorted_df)
```

```python
#6. Vẽ biểu đồ cột cho <AGE>.
plt.bar(df['Name'], df['AGE'])
plt.xlabel('Name')
plt.ylabel('Age')
plt.title('Age of Individuals')
plt.xticks(rotation=90)
plt.show()
```

```python
#7. Vẽ biểu đồ phân tán cho <AGE> và <Height>.
plt.scatter(df['AGE'], df['Height'])
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Age vs Height')
plt.show()
```

```python
#8. Vẽ biểu đồ đường cho cột <Weight>.
plt.plot(df['Name'], df['Weight'])
plt.xlabel('Name')
plt.ylabel('Weight')
plt.title('Weight of Individuals')
plt.xticks(rotation=90)
plt.show()
```

```python
#9. Vẽ biểu đồ tròn cho cột <Height>.
plt.pie(df['Height'], labels=df['Name'], autopct='%1.1f%%', startangle=90)
plt.title('Height Distribution')
plt.show()
```

```python
#10. Vẽ biểu đồ phân tán cho cột <Weight>, <AGE>.
plt.scatter(df['Weight'], df['AGE'])
plt.xlabel('Weight')
plt.ylabel('Age')
plt.title('Weight vs Age')
plt.show()
```

```python
#11. Kiểm tra có giá trị NaN nào trong DataFrame không.
has_nan = df.isna().any().any()
print(f"Có NaN trong DataFrame: {has_nan}")
```

```python
#12. Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên.
df_no_duplicates = df.drop_duplicates(subset=['Name'], keep='first')
print(df_no_duplicates)
```

```python
# 13. Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó.
df['AGE'] = np.where(df['AGE'] > 40, mean_age, df['AGE'])
print(df)
```

```python
# 14. Tính chỉ số BMI và tạo cột <cs_BMI>.
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
print(df)
```

```python
#15. Tính chỉ số BMI và phân loại BMI theo thang đo châu Á
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)
def classify_bmi(bmi):
    if bmi < 18.5:
        return 0  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return 1  # Bình thường
    elif 23 <= bmi <= 24.9:
        return 2  # Thừa cân
    else:
        return 3  # Béo phì

df['Phan_Loai'] = df['cs_BMI'].apply(classify_bmi)
print("DataFrame sau khi tính toán BMI và phân loại:")
print(df[['Name', 'Height', 'Weight', 'cs_BMI', 'Phan_Loai']])
```

```python
# 16. Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị từ 0 tới 1
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df['cs_BMI_normalized'] = scaler.fit_transform(df[['cs_BMI']])
print("\nDataFrame sau khi chuẩn hóa cột 'cs_BMI':")
print(df[['Name', 'cs_BMI', 'cs_BMI_normalized']])
```

```python
#17. Tạo cột giới tính <GT> và loại bỏ từ 'Mr', 'Mrs', 'Miss' trong cột <Name>
def assign_gender(name):
    if 'Mr' in name:
        return 'male'
    else:
        return 'female'

df['GT'] = df['Name'].apply(assign_gender)
df['Name'] = df['Name'].str.replace(r'Mr\.|Mrs\.|Miss\.', '', regex=True).str.strip()
print("\n")
print(df[['Name', 'GT']])
```

```python
# 18. Vẽ biểu đồ tròn cho <GT>
gender_counts = df['GT'].value_counts()
plt.figure(figsize=(6,6))
gender_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set2')
plt.title("Distribution of Gender")
plt.ylabel('')
plt.show()
```

```python
# 19. Trực quan hóa dữ liệu người nữ trên <Phan_Loai>
female_data = df[df['GT'] == 'female']
plt.figure(figsize=(8,6))
female_data['Phan_Loai'].value_counts().plot(kind='bar', color='skyblue')
plt.title("BMI Classification for Females")
plt.xlabel('Phan_Loai (BMI Classification)')
plt.ylabel('Count')
plt.xticks([0, 1, 2, 3], ['Underweight (0)', 'Normal (1)', 'Overweight (2)', 'Obesity (3)'], rotation=45)
plt.show()
```
